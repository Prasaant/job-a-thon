```
train= spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/shared_uploads/prasaanth0x100@gmail.com/train_BRCpofr.csv")
train=train.toPandas()
```

```
train_copy=train.copy()
```

```
train.isnull().sum()
```

```
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```
train=pd.get_dummies(train, columns=['gender'])
train=pd.get_dummies(train, columns=['area'])
train=pd.get_dummies(train, columns=['qualification'])
train=pd.get_dummies(train, columns=['income'])
train=pd.get_dummies(train, columns=['num_policies'])
train=pd.get_dummies(train, columns=['type_of_policy'])
```

```
train=pd.get_dummies(train, columns=['policy'])
```

```
train.columns
```

```
train.info()
```

```
train = train.astype(int)
```

```

train= train.drop(['id'], axis =1)

train= train.drop(['vintage'], axis =1)


```

```
train.corr()
```

```
y=train['cltv']
X= train.drop(['cltv'], axis =1)
X_train, X_test,y_train, y_test = train_test_split(X,y ,
                                   random_state=0, 
                                   test_size=0.2)
```

```
from sklearn.linear_model import LinearRegression
reg = LinearRegression() 
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
```

```
from sklearn import metrics

print("Mean Absolute  Error",metrics.mean_absolute_error(y_test,y_pred))
print("Mean Squared Error",metrics.mean_squared_error(y_test,y_pred))
print("Root Mean Square Error",np.sqrt(metrics.mean_absolute_error(y_test,y_pred)))
```

```
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

Choosing gradient boosting regressor since   large number of environmental variables compared to the number of observations

```
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
GBR = GradientBoostingRegressor()
parameters = {'learning_rate': [0.04,0.5,0.6],
                  'subsample'    : [0.6, 0.5, 0.4],
                  'n_estimators' : [50,100,150],
                  'max_depth'    : [1,2,4] 
                 } 
grid_GBR = GridSearchCV(estimator=GBR, param_grid = parameters, cv = 2, n_jobs=-1)
grid_GBR.fit(X_train, y_train)
```

```
print(" Results from Grid Search " )
print("\n The best estimator across ALL searched params:\n"  ,grid_GBR.best_estimator_)
print("\n The best score across ALL searched params:\n",grid_GBR.best_score_)
print("\n The best parameters across ALL searched params:\n",grid_GBR.best_params_)
```

```

grid_boost = GradientBoostingRegressor(random_state=0,learning_rate= 0.04, max_depth=4, n_estimators= 100, subsample= 0.6)
grid_boost.fit(X_train, y_train)
y_pred = grid_boost.predict(X_test)
score = r2_score(y_test,y_pred)

print("r_2 Score",score)
```

```

print("Mean Absolute  Error",metrics.mean_absolute_error(y_test,y_pred))
print("Mean Squared Error",metrics.mean_squared_error(y_test,y_pred))
print("Root Mean Square Error",np.sqrt(metrics.mean_absolute_error(y_test,y_pred)))
```
